```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import matplotlib.ticker as plticker


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
clinical_trial_data.head()
# Combine the data into a single dataset
merge_table = pd.merge(clinical_trial_data,mouse_drug_data , on="Mouse ID", how="left")
merge_table
# Display the data table for preview
merge_table.head()
```

## Tumor Response to Treatment

```python
# # Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Drug_timepoint = merge_table.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
df_Drug_timepoint =pd.DataFrame(Drug_timepoint).reset_index()

# Preview DataFrame
df_Drug_timepoint.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
SEM_Drug_time_tumor=merge_table.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
df_tumor_vol_errors=pd.DataFrame({"Tumor Volume(mm3) SEM": SEM_Drug_time_tumor}).reset_index()

# # Preview DataFrame
df_tumor_vol_errors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_transposed=df_Drug_timepoint.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
df_transposed.head(15)
```

```python
# Generate the Plot (with Error Bars) 'Capomulin','Infubinol','Ketapril','Placebo'; ''Ceftamin', 'Naftisol','Propriva','Ramicane', 'Stelasyn','Zoniferol'
df_Drug_Timepoint_index=df_tumor_vol_errors.set_index(["Drug"])

error_SEM1 = df_Drug_Timepoint_index["Tumor Volume(mm3) SEM"]["Capomulin"]
leg1=plt.errorbar(df_transposed.index.values, df_transposed["Capomulin"], yerr=error_SEM1, fmt="s", linestyle="-", linewidth=1, alpha=2, capsize=3)

error_SEM2 = df_Drug_Timepoint_index["Tumor Volume(mm3) SEM"]["Infubinol"]
leg2=plt.errorbar(df_transposed.index.values,df_transposed["Infubinol"], yerr=error_SEM2, fmt="o", linestyle="-", linewidth=1, alpha=2, capsize=3)

error_SEM3 = df_Drug_Timepoint_index["Tumor Volume(mm3) SEM"]["Ketapril"]
leg3=plt.errorbar(df_transposed.index.values, df_transposed["Ketapril"], yerr=error_SEM3, fmt="^", linestyle="-", linewidth=1, alpha=2, capsize=3)

error_SEM4 = df_Drug_Timepoint_index["Tumor Volume(mm3) SEM"]["Placebo"]
leg4=plt.errorbar(df_transposed.index.values, df_transposed["Placebo"], yerr=error_SEM4, fmt="D", linestyle="-", linewidth=1, alpha=2, capsize=3)

# # Save the Figure
plt.title("Tumor Response to Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
plt.xlim(-2,50)
plt.ylim(30,75)
plt.legend(handles=[leg1,leg2,leg3,leg4], loc="best")
plt.savefig("Tumor Response To Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
Drug_timepoint_meta = merge_table.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
df_Drug_timepoint_meta =pd.DataFrame(Drug_timepoint_meta).reset_index()

# Preview DataFrame
df_Drug_timepoint_meta.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
SEM_Drug_time_meta=merge_table.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
df_meta_vol_errors=pd.DataFrame({"Metastatic Sites SEM": SEM_Drug_time_meta}).reset_index()

# Preview DataFrame

df_meta_vol_errors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_transposed_meta=df_Drug_timepoint_meta.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
df_transposed_meta.head(5)
```

```python
# Generate the Plot (with Error Bars) 'Capomulin','Infubinol','Ketapril','Placebo'; ''Ceftamin', 'Naftisol','Propriva','Ramicane', 'Stelasyn','Zoniferol'
df_Drug_Timepoint_index=df_meta_vol_errors.set_index(["Drug"])


error_SEM1 = df_Drug_Timepoint_index["Metastatic Sites SEM"]["Capomulin"]
leg1=plt.errorbar(df_transposed.index.values,df_transposed_meta["Capomulin"] , yerr=error_SEM1, fmt="s", linestyle="-", linewidth=1, alpha=2, capsize=3)

error_SEM2 = df_Drug_Timepoint_index["Metastatic Sites SEM"]["Infubinol"]
leg2=plt.errorbar(df_transposed.index.values,df_transposed_meta["Infubinol"], yerr=error_SEM2, fmt="o", linestyle="-", linewidth=1, alpha=2, capsize=3)

error_SEM3 = df_Drug_Timepoint_index["Metastatic Sites SEM"]["Ketapril"]
leg3=plt.errorbar(df_transposed.index.values,df_transposed_meta["Ketapril"], yerr=error_SEM3, fmt="^", linestyle="-", linewidth=1, alpha=2, capsize=3)

error_SEM4 = df_Drug_Timepoint_index["Metastatic Sites SEM"]["Placebo"]
leg4=plt.errorbar(df_transposed.index.values,df_transposed_meta["Placebo"], yerr=error_SEM4, fmt="D", linestyle="-", linewidth=1, alpha=2, capsize=3)
# # Save the Figure
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Meta. Sites")
plt.grid()
plt.xlim(-2,50)
plt.ylim(-.1,4)
plt.legend(handles=[leg1,leg2,leg3,leg4], loc="best")
plt.savefig("Metastatic Spread During Treatment.png")
```

## Survival Rates

```python


# # Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
Drug_timepoint_survive = merge_table.groupby(["Drug", "Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
df_Drug_timepoint_survive =pd.DataFrame(Drug_timepoint_survive).reset_index()

# Preview DataFrame
df_Drug_timepoint_survive.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_transposed_survive=df_Drug_timepoint_survive.pivot(index='Timepoint', columns='Drug', values='Mouse ID')

# Preview the Data Frame
df_transposed_survive.head()
```

```python
# Generate the Plot (Accounting for percentages)
leg1=plt.plot(df_transposed_survive.index.values,df_transposed_survive["Capomulin"]/25*100 , marker="s",linestyle="-",linewidth=1, alpha=2, label="Capomulin")
leg2=plt.plot(df_transposed_survive.index.values,df_transposed_survive["Infubinol"]/25*100 , marker="o",linestyle="-",linewidth=1, alpha=2,label ="Infubinol")
leg3=plt.plot(df_transposed_survive.index.values,df_transposed_survive["Ketapril"]/25*100 , marker="^",linestyle="-",linewidth=1, alpha=2,label ="Ketapril")
leg4=plt.plot(df_transposed_survive.index.values,df_transposed_survive["Placebo"]/25*100 , marker="D",linestyle="-",linewidth=1, alpha=2,label ="Placebo")


# # Save the Figure
plt.title("Survival During Treatment")
plt.xlabel("Time(Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()
# plt.xlim(-2,50)
# plt.ylim(-.1,4)
plt.legend(loc='lower left')
plt.savefig("Survival During Treatment.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
df_transposed_loc=df_transposed.iloc[[0,-1]]
df_transposed_diff=df_transposed_loc.diff()/45*100
df_transposed_diff
df_transposed_diff=df_transposed_diff.iloc[1:]
df_transposed_diff=df_transposed_diff.melt()
df_transposed_diff=df_transposed_diff.rename(columns = {'value':'Percentage Change'})
df_transposed_diff.describe
df_transposed_diff.head(15)
# Display the data to confirm
```

```python

df_transposed_diff=df_transposed_diff.drop(df_transposed_diff.index[[1,4,6,7,8,9]])
```

```python

df_transposed_diff["Green"]=df_transposed_diff["Percentage Change"]>0

fig, ax1 = plt.subplots()

bar_chart=plt.bar(df_transposed_diff["Drug"], df_transposed_diff["Percentage Change"],width=-1.0,align='edge',color=df_transposed_diff.Green.map({True: 'red', False: 'green'}))
plt.title("Tumor Change Over a 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        
        height = rect.get_height()
        if height>0:
            ax1.text(rect.get_x() + rect.get_width()/2., height/height,
                '%d%%' % int(height),
                ha='center', va='bottom', color="w")
        if height<0:
            ax1.text(rect.get_x() + rect.get_width()/2., abs(height)/height-4.5,
                '%d%%' % int(height),
                ha='center', va='bottom', color="w")

#Spacing between each line
spacing =20
loc = plticker.MultipleLocator(spacing)
ax1.yaxis.set_major_locator(loc)


# Add the grid
ax1.grid(which='major', axis='both', linestyle='-')

autolabel(bar_chart)
plt.ylim(-25,70)
# Show the Figure
plt.savefig("Summary Bar Graph.png")
fig.show()
```
